## 1. What are Data Types?
In Python, every value has a type. The four basic types are:
- **Integer:** Whole numbers like 22, 30, 100
- **Float:** Decimal numbers like 3.14, 3.85
- **Boolean:** Only True or False
- **String:** Any text written inside quotes

```python
# Integer examples
age = 22
total_days = 30

# Float examples
gpa = 3.85
pi_value = 3.14159

# Boolean examples
is_intern = True
is_complete = False

# String examples
intern_name = "Sehar Andleeb"
company = "Xeven Solutions"

# Check all types
print(type(age))
print(type(gpa))
print(type(is_intern))
print(type(intern_name))
```

## 2. Variable Naming — PEP 8 Standards
- Always use snake_case for variable names
- Names should be descriptive and meaningful
- Constants are written in UPPERCASE_WITH_UNDERSCORES

```python
# Correct PEP 8 naming
intern_name = "Sehar Andleeb"
current_day = 2
MAX_DAYS = 30

# Dynamic typing — variable can change type
my_variable = 10
print(type(my_variable))

my_variable = "Now I am a string"
print(type(my_variable))
```

## 3. Input and Output
- print() displays output to the screen
- input() takes text from the user
- input() always returns a string
- f-strings let us put variables inside text cleanly

```python
# print() with f-string
intern_name = "Sehar Andleeb"
current_day = 2

print(f"My name is {intern_name}")
print(f"Today is Day {current_day} of 30")

# print() with separator
print("Python", "AI", "Internship", sep=" | ")
```

## 4. Type Conversion
Converting one data type to another:
- int() converts to integer
- float() converts to float
- str() converts to string
- bool() converts to boolean

Important: int() on a float removes the decimal, it does not round

```python
# String to integer
age = int("22")
print(age, type(age))

# String to float
gpa = float("3.85")
print(gpa, type(gpa))

# Float to integer — decimal is removed
pi = int(3.99)
print(pi)

# Boolean conversions
print(int(True))   # 1
print(int(False))  # 0
print(bool(""))    # False
print(bool("hi"))  # True
```

## 5. Troubleshooting Notes

| Problem | Cause | Fix |
|---|---|---|
| ValueError: invalid literal for int() | Converting non-numeric string to int | Make sure string contains only numbers |
| input() returns string not number | input() always returns str | Wrap with int() or float() |
| NameError: name is not defined | Variable used before assignment | Always assign before using |

## 6. Key Learnings

- Python has 4 basic data types: int, float, bool, string
- Variables are dynamically typed — they can change type anytime
- input() always returns a string, convert it when needed
- PEP 8 requires snake_case for variables and UPPERCASE for constants
- int() on a float removes the decimal part, it does not round
- True equals 1 and False equals 0 in Python
