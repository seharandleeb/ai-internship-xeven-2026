## 1. Theoretical Concepts

### 1.1 Function Definition
A function is a named, reusable block of code that does one task. The `def` keyword starts the definition; **parameters** are placeholders in the definition while **arguments** are the actual values passed in the call. `return` hands a value back to the caller. A function with no `return` (or a bare `return`) returns `None` — `print()` only *displays*, it does not hand a value back.

### 1.2 Arguments
- **Positional** — matched by order: `power(2, 3)`.
- **Keyword** — matched by name: `power(exp=3, base=2)`.
- **Default** — optional with a fallback: `greet(name, greeting="Hello")`. Never use a *mutable* default (`items=[]`); use `None` and create inside.
- **Unpacking** — `*args` collects extra positional args into a tuple, `**kwargs` collects extra keyword args into a dict.

### 1.3 Scope
Python resolves names with the **LEGB** rule: Local → Enclosing → Global → Built-in. Locals exist only inside their function and die when it returns; globals live for the program's lifetime. Reading a global inside a function is fine, but **reassigning** one requires the `global` keyword (use sparingly — prefer params and returns).

### 1.4 Function Design Principles
- **Single Responsibility** — one function, one job.
- **Descriptive names** — `get_standard_deviation()` not `gsd()`; `is_`/`validate_` prefixes for booleans.
- **Docstrings** — every function documents its purpose, `Args`, `Returns`, `Raises`, and an `Example`.


## 2. Task 1 — Math Utility Library

Functions with parameter validation (empty lists, non-numeric values), default `precision`, and an optional `sample` flag for sample vs population standard deviation.

```python
from math_utils import (
    calculate_average,
    find_median,
    get_standard_deviation,
)

data = [10, 20, 30, 40, 50]
print('Average:', calculate_average(data))
print('Median: ', find_median(data))
print('Std Dev (sample):', get_standard_deviation(data))
print('Std Dev (pop):   ',
      get_standard_deviation(data, sample=False))
```

**Validation in action** — bad input raises clear errors:

```python
for bad in ([], [1, 'two', 3]):
    try:
        calculate_average(bad)
    except (ValueError, TypeError) as err:
        print(f'{bad!r:18} -> {type(err).__name__}: {err}')
```

## 3. Task 2 — Text Processing Functions

Small single-purpose functions chained inside `process_text()`, which returns a tuple `(processed_text, word_count, char_count, unique_words)`.

```python
from text_processing import (
    count_words,
    extract_emails,
    remove_punctuation,
    title_case,
    process_text,
)

sample = 'Hello, world! Email me at umar@xeven.com please.'
print('Word count:', count_words(sample))
print('Emails:    ', extract_emails(sample))
print('No punct:  ', remove_punctuation(sample))
print('Title case:', title_case(sample))
```

**Chaining + multiple return values via tuple unpacking:**

```python
processed, words, chars, unique = process_text(sample)
print('Processed:   ', processed)
print('Word count:  ', words)
print('Char count:  ', chars)
print('Unique words:', unique)

# Flags toggle the chained steps:
raw = process_text(sample, remove_punct=False,
                   to_title=False)
print('No cleaning: ', raw[0])
```

## 4. Task 3 — Validation Function Suite

Each validator returns a `(is_valid: bool, error_message: str)` tuple. `validate_user_input()` is a wrapper that dispatches to the correct validator by field type.

```python
from validators import validate_user_input

checks = [
    ('email', 'umar@xeven.com'),
    ('email', 'bad-email'),
    ('phone', '+1 (555) 123-4567'),
    ('date', '2026-06-12'),
    ('date', '2026-13-99'),
    ('password', 'Str0ng!Pass'),
    ('password', 'weak'),
]
for field, val in checks:
    ok, msg = validate_user_input(field, val)
    status = 'PASS' if ok else f'FAIL ({msg})'
    print(f'{field:9} {val:22} -> {status}')
```

**Unsupported field type raises `ValueError`** (fail-fast design):

```python
try:
    validate_user_input('username', 'umar123')
except ValueError as err:
    print('ValueError:', err)
```
