## 1. What are Lists?

Lists are ordered mutable collections that can hold multiple
items of different data types. They are one of the most
commonly used data structures in Python.

### Key Properties
- **Ordered** — items maintain their insertion order
- **Mutable** — items can be added, removed, or changed
- **Indexed** — each item has a position starting from 0
- **Mixed types** — can hold integers, strings, booleans together

```python
# List creation examples

empty_list  = []                           # empty list
numbers     = [1, 2, 3, 4, 5]             # integer list
fruits      = ["apple", "banana", "mango"] # string list
mixed       = [1, "hello", 3.14, True]    # mixed types
nested      = [[1, 2], [3, 4], [5, 6]]    # nested list

print(f"Empty    : {empty_list}")
print(f"Numbers  : {numbers}")
print(f"Fruits   : {fruits}")
print(f"Mixed    : {mixed}")
print(f"Nested   : {nested}")
print(f"Length   : {len(fruits)}")
```

## 2. Indexing and Slicing

### Positive Indexing
Starts from 0 at the beginning of the list.

### Negative Indexing
Starts from -1 at the end of the list.

### Slicing Syntax
list[start:end:step]
- start — where to begin (inclusive)
- end — where to stop (exclusive)
- step — how many items to skip

```python
fruits = ["apple", "banana", "cherry", "mango", "orange"]

# Positive indexing
print(f"fruits[0]    : {fruits[0]}")
print(f"fruits[2]    : {fruits[2]}")

# Negative indexing
print(f"fruits[-1]   : {fruits[-1]}")
print(f"fruits[-2]   : {fruits[-2]}")

# Slicing
numbers = list(range(1, 11))
print(f"\nOriginal     : {numbers}")
print(f"First 3      : {numbers[:3]}")
print(f"Last 3       : {numbers[-3:]}")
print(f"Every 2nd    : {numbers[::2]}")
print(f"Reversed     : {numbers[::-1]}")
print(f"Middle       : {numbers[3:7]}")
```

## 3. Adding and Removing Items

| Method | Description | Example |
|---|---|---|
| append() | Add one item at end | list.append(5) |
| insert() | Add at specific index | list.insert(2, 5) |
| extend() | Add multiple items | list.extend([4,5,6]) |
| remove() | Remove by value | list.remove("apple") |
| pop() | Remove by index | list.pop(0) |
| clear() | Remove everything | list.clear() |

```python
fruits = ["apple", "banana", "cherry"]
print(f"Original     : {fruits}")

# Adding items
fruits.append("mango")
print(f"append()     : {fruits}")

fruits.insert(1, "orange")
print(f"insert()     : {fruits}")

fruits.extend(["grape", "kiwi"])
print(f"extend()     : {fruits}")

# Removing items
fruits.remove("banana")
print(f"remove()     : {fruits}")

popped = fruits.pop(0)
print(f"pop(0)       : {fruits} — removed {popped}")
```

## 4. List Methods

| Method | Description |
|---|---|
| sort() | Sort in place — modifies original |
| sorted() | Returns new sorted list |
| reverse() | Reverse in place |
| count() | Count occurrences of value |
| index() | Find index of first occurrence |
| len() | Get total number of items |

```python
numbers = [5, 2, 8, 1, 9, 3, 7, 2, 2]
print(f"Original     : {numbers}")

# count and index
print(f"count(2)     : {numbers.count(2)}")
print(f"index(9)     : {numbers.index(9)}")

# sort vs sorted
original    = [5, 2, 8, 1, 9]
new_sorted  = sorted(original)
print(f"\noriginal     : {original}")
print(f"sorted()     : {new_sorted}")
print(f"original     : {original} — unchanged")

original.sort()
print(f"sort()       : {original} — changed in place")
```

## 5. List Comprehension

A clean one-line way to create lists. Faster and more
Pythonic than regular for loops.

### Syntax
[expression for item in iterable if condition]

```python
# Basic comprehension
squares = [x ** 2 for x in range(1, 6)]
print(f"Squares      : {squares}")

# With condition
evens = [x for x in range(1, 11) if x % 2 == 0]
print(f"Even numbers : {evens}")

# String transformation
names = ["sehar", "ali", "sara"]
upper = [name.upper() for name in names]
print(f"Uppercase    : {upper}")

# Filter passing marks
marks  = [45, 78, 92, 38, 65, 88, 23, 71]
passed = [m for m in marks if m >= 50]
print(f"All marks    : {marks}")
print(f"Passed only  : {passed}")
```

## 6. Error Handling

Common list errors:
- **IndexError** : accessing index that does not exist
- **ValueError** : removing value that is not in list

```python
numbers = [10, 20, 30, 40, 50]

# IndexError handling
try:
    value = numbers[10]
except IndexError as error:
    print(f"IndexError   : {error}")
    print(f"Valid range  : 0 to {len(numbers) - 1}")

# ValueError handling
try:
    numbers.remove(999)
except ValueError as error:
    print(f"ValueError   : {error}")

# Safe access function
def safe_get(lst, index, default=None):
    """Safely get item without IndexError."""
    try:
        return lst[index]
    except IndexError:
        return default

print(f"Safe index 2 : {safe_get(numbers, 2)}")
print(f"Safe index 10: {safe_get(numbers, 10, 'Not found')}")
```

## 8. Key Learnings

- Lists are ordered, mutable, and can hold mixed data types
- Indexing starts at 0, negative indexing starts at -1
- Slicing uses start:end:step — end is never included
- append() adds one item, extend() adds many, insert() at position
- remove() deletes by value, pop() by index, clear() removes all
- sort() modifies original, sorted() returns new list
- List comprehension is faster and cleaner than for loops
- IndexError and ValueError must be handled with try-except
- Parallel lists use same index to represent related data
