## 1. Arithmetic Operators

```python
# All arithmetic operators in Python
a = 17
b = 5

print("-- Arithmetic Operators --")
print(f"  {a} + {b}  = {a + b}")    # Addition
print(f"  {a} - {b}  = {a - b}")    # Subtraction
print(f"  {a} * {b}  = {a * b}")    # Multiplication
print(f"  {a} / {b}  = {a / b}")    # True division → always float
print(f"  {a} // {b} = {a // b}")   # Floor division → rounds down
print(f"  {a} % {b}  = {a % b}")    # Modulo → remainder
print(f"  {a} ** {b} = {a ** b}")   # Exponentiation
```

## 2. Operator Precedence (PEMDAS)
Python evaluates expressions in this order:
1. `()` Parentheses
2. `**` Exponentiation
3. `*, /, //, %` Multiplication/Division
4. `+, -` Addition/Subtraction

```python
# Demonstrating precedence — same numbers, different results
print("-- Operator Precedence --")
print(f"  2 + 3 * 4      = {2 + 3 * 4}")       # 14 — multiply first
print(f"  (2 + 3) * 4    = {(2 + 3) * 4}")     # 20 — parentheses first
print(f"  2 ** 3 + 1     = {2 ** 3 + 1}")       # 9  — exponent first
print(f"  10 - 4 / 2     = {10 - 4 / 2}")       # 8.0 — division first

print()
print("-- Complex Expression Breakdown --")
# 100 - 2 ** 3 * 4 + 10 / 2
# Step 1: 2**3 = 8
# Step 2: 8*4  = 32
# Step 3: 10/2 = 5.0
# Step 4: 100 - 32 = 68
# Step 5: 68 + 5.0 = 73.0
result = 100 - 2 ** 3 * 4 + 10 / 2
print(f"  100 - 2**3 * 4 + 10/2 = {result}")   # 73.0
```

## 3. Exponent Associativity
Unlike other operators, `**` evaluates **right to left**.

```python
# Exponent is RIGHT-associative
print("-- Exponent Associativity --")
print(f"  2 ** 3 ** 2   = {2 ** 3 ** 2}")      # 512 → 2 ** (3**2) = 2**9
print(f"  (2 ** 3) ** 2 = {(2 ** 3) ** 2}")    # 64  → (2**3)**2   = 8**2

print()
# Practical use of exponents
print("-- Practical Exponent Uses --")
print(f"  Square root of 16  : {16 ** 0.5}")   # 4.0
print(f"  Cube root of 27    : {27 ** (1/3):.4f}")  # 3.0
print(f"  2 to the power 10  : {2 ** 10}")     # 1024
```

## 4. Floor Division and Modulo
These two operators work as a pair — very useful in real programs.

```python
# Floor division always rounds DOWN
print("-- Floor Division --")
print(f"  7 // 2  = {7 // 2}")      # 3  (not 3.5)
print(f"  -7 // 2 = {-7 // 2}")     # -4 (rounds DOWN, not toward zero!)
print(f"  7.5 // 2 = {7.5 // 2}")   # 3.0 (float floor division)

print()
# Modulo gives the remainder
print("-- Modulo Use Cases --")
number = 47
print(f"  {number} is even?          {number % 2 == 0}")
print(f"  {number} divisible by 5?   {number % 5 == 0}")
print(f"  Last digit of {number}:     {number % 10}")

# Real use: convert minutes to hours and minutes
total_minutes = 135
hours = total_minutes // 60
minutes = total_minutes % 60
print(f"\n  {total_minutes} minutes = {hours} hours and {minutes} minutes")
```

## 5. Comparison and Logical Operators

```python
age = 22
has_id = True
is_student = True

print("-- Comparison Operators --")
print(f"  age == 22        : {age == 22}")    # True  — equal to
print(f"  age != 18        : {age != 18}")    # True  — not equal
print(f"  age > 18         : {age > 18}")     # True  — greater than
print(f"  age <= 22        : {age <= 22}")    # True  — less or equal
print(f"  18 <= age <= 30  : {18 <= age <= 30}")  # True — chained

print()
print("-- Logical Operators --")
print(f"  age>=18 and has_id          : {age >= 18 and has_id}")   # True
print(f"  is_student or age > 25      : {is_student or age > 25}") # True
print(f"  not is_student              : {not is_student}")          # False
print(f"  age>=18 and has_id and not (age>65) : {age>=18 and has_id and not (age>65)}")
```

## 6. Type Conversion — Implicit vs Explicit

```python
print("-- Implicit Conversion (Python does it automatically) --")
x = 10       # int
y = 3.5      # float
result = x + y
print(f"  int + float = {result}  (type: {type(result).__name__})")  # float

print()
print("-- Explicit Conversion --")
print(f"  int(3.9)     = {int(3.9)}")        # 3   — truncates, does NOT round
print(f"  int('42')    = {int('42')}")        # 42
print(f"  int(True)    = {int(True)}")        # 1
print(f"  float(7)     = {float(7)}")         # 7.0
print(f"  float('3.14')= {float('3.14')}")    # 3.14
print(f"  str(100)     = '{str(100)}'")       # '100'
print(f"  bool(0)      = {bool(0)}")          # False
print(f"  bool(1)      = {bool(1)}")          # True
print(f"  bool('')     = {bool('')}")         # False
print(f"  bool('hi')   = {bool('hi')}")       # True
```

## 7. Safe Type Conversion with try/except
Always use `try/except` when converting user input — it prevents crashes.

```python
def safe_convert(value: str) -> None:
    """Attempt int conversion and report success or failure."""
    try:
        result = int(value)
        print(f"  int('{value}') -> SUCCESS : {result}")
    except ValueError as e:
        print(f"  int('{value}') -> FAILED  : {e}")

print("-- Safe Conversion Examples --")
safe_convert("42")       # works
safe_convert("3.14")     # fails — can't convert float string to int directly
safe_convert("hello")    # fails — not a number
safe_convert("")         # fails — empty string
safe_convert("  7  ")    # works — int() handles whitespace
```

## Summary — Key Takeaways

| Concept | Key Point |
|---------|-----------|
| Precedence | PEMDAS — use `()` when in doubt |
| `**` | Right-associative — evaluates right to left |
| `//` | Always rounds DOWN (even for negatives) |
| `%` | Gives remainder — useful for even/odd, last digit |
| `int()` | Truncates — `int(3.9)` = `3` not `4` |
| `input()` | Always returns `str` — convert before math |
| `try/except` | Always wrap conversions of user input |
| Logical ops | `and` needs ALL true, `or` needs ONE true |
