## AI Tool Research Comparison

**Research Question:** What is RAG and how does LangChain implement it?

| Aspect | ChatGPT  | Gemini  | Claude  |
|--------|-------------------|------------------|-----------------|
| RAG Definition | Combines retrieval + generation to ground LLM answers in documents (confirm) | Retrieves relevant context from a knowledge base before generating a response (confirm) | RAG indexes documents into a vector store, retrieves top-k chunks for a query, passes them to an LLM as context to generate a grounded answer |
| LangChain RetrievalQA | Wraps retriever + LLM into a single chain that handles retrieval and generation (confirm) | Provides a ready-made chain that connects a vector store retriever to an LLM (confirm) | RetrievalQA.from_chain_type() accepts a retriever, LLM, and optional prompt template. Returns source_documents when return_source_documents=True |
| FAISS Role | Fast approximate nearest neighbor search for vector similarity (confirm) | Stores document embeddings and retrieves most similar vectors for a query (confirm) | FAISS.from_documents() embeds and indexes chunks. as_retriever(search_kwargs={'k':4}) returns top-4 similar chunks per query |
| Custom Prompts | PromptTemplate lets you define how context and question are presented to the LLM (confirm) | Allows strict or detailed instruction modes to control LLM behavior (confirm) | PromptTemplate with input_variables=['context','question'] controls tone, strictness, and citation behavior of the LLM response |
| Hallucination Reduction | RAG grounds answers in retrieved documents, reducing made-up facts (confirm) | Providing context limits the LLM to known information, reducing hallucinations (confirm) | Strict prompt ('Answer ONLY from context') combined with retrieval forces the LLM to refuse out-of-context queries rather than hallucinate |
| Multi-Doc RAG | Metadata filtering allows searching specific document subsets (confirm) | Tagging documents by source type enables filtered retrieval (confirm) | Each Document gets source_type metadata. Filtered FAISS stores per source_type enable targeted retrieval. Synthesis combines chunks across all source types |

## Key Learnings

1. **RAG reduces hallucinations** by grounding LLM answers in retrieved document context
2. **Custom prompt templates** control LLM behavior — strict mode forces refusal of out-of-context queries
3. **Metadata tagging** enables filtered retrieval across multiple document source types
4. **FAISS indexing** is extremely fast — 12 chunks indexed in 189ms
5. **Version pinning** is critical — langchain 0.3.0 + langchain-core 0.3.0 must match
6. **DLL issues** with PyTorch on Windows can be bypassed using offline embedders for development
