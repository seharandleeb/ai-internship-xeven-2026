## What is Artificial Intelligence?

After reading multiple sources, here is my understanding of key concepts:

### Definition
AI is the ability of a computer to learn from data and make decisions
without being explicitly programmed for every situation.

### Narrow AI vs General AI
- **Narrow AI:** Designed for one specific task only
  - Example: Google Translate, Face ID, spam filter
- **General AI:** Can perform any intellectual task a human can
  - Example: Does not exist yet, still a future concept

### Real World Applications
- **Healthcare:** AI detects cancer in X-rays faster than doctors
- **Finance:** AI flags suspicious transactions to prevent fraud
- **Education:** AI personalizes lessons based on student progress
- **Marketing:** AI recommends products based on your behavior

### Key AI Fields
- **Machine Learning:** System learns patterns from data
- **Deep Learning:** Uses neural networks for complex tasks like images
- **NLP:** Helps computers understand human language (ChatGPT)
- **Computer Vision:** Helps computers understand images (Face ID)

### Training vs Inference
- **Training:** Feeding data to the model so it learns patterns
- **Inference:** Using the trained model to make predictions on new data
