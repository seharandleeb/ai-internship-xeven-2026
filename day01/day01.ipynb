## 5. Troubleshooting Notes

| Problem | Cause | Fix |
|---|---|---|
| source .venv/bin/activate not working | Linux command, not valid on Windows | Used .venv\Scripts\activate instead |
| fatal: not a git repository | git init not run in root folder | Ran git init in correct folder |
| day01/ does not have a commit checked out | Nested .git folder inside day01 | Removed using rmdir /s /q day01\.git |
| src refspec main does not match any | No commit made before push | Made commit first then pushed |

## 6. Key Learnings

- Virtual environments keep project dependencies isolated and clean
- Always run git init in the root folder, never inside subfolders
- .gitignore should be created before the first commit
- On Windows, use .venv\Scripts\activate instead of source command
