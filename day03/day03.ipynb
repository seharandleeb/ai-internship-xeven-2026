## 1. What are Conditional Statements?

Conditional statements allow a program to make decisions based on
whether a condition is True or False. Instead of running every line
top to bottom, the program chooses which path to follow.

### Syntax
- **if** — runs when condition is True
- **elif** — checks another condition if previous was False
- **else** — runs when all conditions above are False

### Key Rule
Only ONE block runs in an if-elif-else chain.

```python
# Basic if-elif-else demonstration
# Comparison operators: ==, !=, >, <, >=, <=

marks = 85

# Each elif checks the next condition only if previous was False
if marks >= 90:
    print("Grade: A — Excellent")
elif marks >= 80:
    print("Grade: B — Good")
elif marks >= 70:
    print("Grade: C — Average")
elif marks >= 60:
    print("Grade: D — Below Average")
else:
    print("Grade: F — Fail")
```

## 2. Boolean Logic and Truthy/Falsy Values

In Python, every value is either Truthy or Falsy.

| Falsy Values | Truthy Values |
|---|---|
| False | True |
| 0 | Any non-zero number |
| "" (empty string) | Any non-empty string |
| None | Any list, dict, or object |

```python
# Truthy and Falsy demonstration

values = [0, 1, "", "hello", None, True, False, [], [1, 2]]

for value in values:
    if value:
        print(f"{repr(value):15} → Truthy")
    else:
        print(f"{repr(value):15} → Falsy")
```

## 3. Comparison Operators

| Operator | Meaning | Example | Result |
|---|---|---|---|
| == | Equal to | 5 == 5 | True |
| != | Not equal to | 5 != 3 | True |
| > | Greater than | 7 > 3 | True |
| < | Less than | 2 < 8 | True |
| >= | Greater or equal | 5 >= 5 | True |
| <= | Less or equal | 4 <= 6 | True |

```python
# Comparison operators in action

age = 20

# == checks equality, = assigns a value
print(age == 20)    # True
print(age != 18)    # True
print(age > 18)     # True
print(age < 18)     # False
print(age >= 20)    # True
print(age <= 25)    # True
```

## 4. Nested Conditionals

Nested conditionals are if statements inside other if statements.
Use them when a decision depends on a previous decision being True.

Best practice: avoid more than 2 levels of nesting for readability.

```python
# Nested conditional example
# Outer condition checks category, inner gives specific message

marks = 92
student_name = "Sehar"

if marks >= 80:
    # Outer: marks are in A or B range
    if marks >= 90:
        # Inner: more specific check within the A range
        print(f"Outstanding {student_name}! Grade A+")
    else:
        print(f"Great job {student_name}! Grade A")
else:
    print(f"Keep working {student_name}!")
```

## 5. Common Pitfalls

| Pitfall | Wrong | Correct |
|---|---|---|
| Assignment vs Comparison | if age = 18 | if age == 18 |
| Indentation error | 4 spaces missing | Always use 4 spaces |
| Wrong order in elif | Check < 50 before < 90 | Always check highest first |

```python
# Error handling with conditionals

try:
    user_input = "abc"
    age = int(user_input)

    if age < 0:
        print("Age cannot be negative")
    elif age > 120:
        print("Please enter a realistic age")
    else:
        print(f"Valid age: {age}")

except ValueError:
    print("Invalid input — please enter a number only")
```

## 7. Key Learnings

- if-elif-else allows programs to make decisions based on conditions
- Only ONE block in an if-elif-else chain ever runs
- Always check the highest boundary first in elif chains
- Use == for comparison, never = inside an if statement
- Nested conditionals should not go deeper than 2 levels
- Every value in Python is either Truthy or Falsy
- try-except handles invalid inputs without crashing the program
- Flat elif chains are more readable than deeply nested if statements
