## 1. What is Machine Learning?

Machine Learning is a branch of AI where computers learn from
data without being explicitly programmed for every situation.
Instead of writing rules manually, we feed data to the model
and it finds patterns on its own.

### Three Main Types
- **Supervised Learning** — learns from labeled data
- **Unsupervised Learning** — finds patterns in unlabeled data
- **Reinforcement Learning** — learns through trial and error

## 2. Supervised Learning

The model is trained on labeled data — every input has a
correct output. The model learns the relationship between
input and output and uses it to predict on new data.

### Real World Examples
- Email spam detection
- House price prediction
- Disease diagnosis
- Image recognition

````python
# Supervised Learning Simulation
# Model learns from labeled training data

# Training data — input and correct output pairs
training_data = [
    {"email": "Win a prize now",  "label": "Spam"},
    {"email": "Meeting at 3pm",   "label": "Not Spam"},
    {"email": "Free money click", "label": "Spam"},
    {"email": "Project deadline", "label": "Not Spam"},
]

# Keywords model learned from training data
spam_keywords = ["win", "free", "prize", "money", "claim"]

# Make predictions on new unseen emails
test_emails = ["Win free prizes", "Team lunch tomorrow"]

print("Supervised Learning — Spam Classifier")
print("-" * 40)
for email in test_emails:
    # Check if spam keyword exists in email
    is_spam = any(word in email.lower() for word in spam_keywords)
    prediction = "Spam" if is_spam else "Not Spam"
    print(f"Email     : {email}")
    print(f"Prediction: {prediction}\n")
````

## 3. Unsupervised Learning

No labels are given. The model finds hidden patterns and
groups in the data by itself. Used when we do not know
what categories exist in advance.

### Real World Examples
- Customer segmentation
- Anomaly detection
- Topic modeling
- Recommendation systems

````python
# Unsupervised Learning Simulation
# Model groups customers without any labels

customers = [
    {"name": "Ali",    "spending": 5000},
    {"name": "Sara",   "spending": 25000},
    {"name": "Ahmed",  "spending": 4500},
    {"name": "Fatima", "spending": 30000},
    {"name": "Usman",  "spending": 3000},
]

# Model discovers groups based on spending patterns
high_value = []
low_value  = []

for customer in customers:
    if customer["spending"] >= 20000:
        high_value.append(customer["name"])
    else:
        low_value.append(customer["name"])

print("Unsupervised Learning — Customer Segmentation")
print("-" * 40)
print(f"High Value Customers : {high_value}")
print(f"Low Value Customers  : {low_value}")
print("\nNote: No labels were given — model found groups itself")
````

## 4. Reinforcement Learning

An agent learns by interacting with an environment.
Good actions receive rewards, bad actions receive penalties.
The agent learns to maximize total reward over time.

### Key Components
- **Agent** — the learner making decisions
- **Environment** — what the agent interacts with
- **Action** — what the agent does
- **Reward** — feedback received after each action

### Real World Examples
- Game AI (Chess, Go)
- Robot navigation
- Self-driving cars
- Trading systems

````python
# Reinforcement Learning Simulation
# Agent learns through trial and error

episodes = [
    {"action": "Move Right", "reward": 10},
    {"action": "Move Left",  "reward": -5},
    {"action": "Move Up",    "reward": 50},
    {"action": "Move Down",  "reward": -1},
    {"action": "Move Right", "reward": 10},
]

total_score = 0

print("Reinforcement Learning — Game Agent")
print("-" * 40)
print(f"{'Action':<15} {'Reward':<10} {'Total Score'}")
print(f"{'-'*15} {'-'*10} {'-'*11}")

for episode in episodes:
    total_score += episode["reward"]
    print(f"{episode['action']:<15} "
        f"{episode['reward']:<10} "
        f"{total_score}")

print(f"\nAgent learned: Move Right and Up give rewards!")
````

## 5. Regression vs Classification

| | Regression | Classification |
|---|---|---|
| Output | Continuous number | Discrete category |
| Example | House price = 250,000 | Email = Spam |
| Example | Temperature = 36.6°C | Disease = Yes/No |
| Algorithm | Linear Regression | Decision Tree |

````python
# Regression — predicts continuous value
def predict_house_price(size, bedrooms):
    """Predict house price using simple formula."""
    base_price    = 50000
    size_value    = size * 100
    bedroom_value = bedrooms * 15000
    return base_price + size_value + bedroom_value


# Classification — predicts discrete category
def classify_email(has_spam_words, unknown_sender):
    """Classify email as spam or not spam."""
    spam_score = 0
    if has_spam_words:
        spam_score += 40
    if unknown_sender:
        spam_score += 25
    return "Spam" if spam_score >= 40 else "Not Spam"


# Regression output — continuous number
price = predict_house_price(1500, 3)
print(f"Regression Output    : ${price:,} (continuous number)")

# Classification output — discrete category
result = classify_email(True, True)
print(f"Classification Output: {result} (discrete category)")
````

## 6. Decision Trees

A decision tree makes decisions by asking a series of
questions. Each question splits the data into branches
until a final answer is reached.

### Structure
- **Root Node** — first question at the top
- **Internal Nodes** — questions in the middle branches
- **Leaf Nodes** — final answers at the bottom

### Visual Structure
```
Is age >= 18?
├── No  → Reject
└── Yes → Is income >= 25000?
          ├── No  → Reject
          └── Yes → Approve
```

````python
# Decision Tree Simulation
# Loan approval using branching conditions

def loan_decision_tree(age, income, credit_score):
    """
    Simulate loan approval decision tree.

    Parameters:
        age          (int): Applicant age
        income       (float): Annual income
        credit_score (int): Credit score

    Returns:
        str: Loan decision
    """
    # Root node — age check
    if age < 18:
        return "Rejected — Must be 18 or older"

    # Internal node — credit score check
    elif credit_score < 500:
        return "Rejected — Credit score too low"

    # Internal node — income check
    elif income < 25000:
        return "Rejected — Income too low"

    # Leaf node — approved
    else:
        return "Approved"


# Test the decision tree
test_cases = [
    (16, 30000, 700),
    (25, 20000, 650),
    (30, 50000, 400),
    (35, 60000, 720),
]

print("Decision Tree — Loan Approval")
print("-" * 45)
print(f"{'Age':<6} {'Income':<10} {'Score':<8} {'Decision'}")
print(f"{'-'*6} {'-'*10} {'-'*8} {'-'*20}")

for age, income, score in test_cases:
    decision = loan_decision_tree(age, income, score)
    print(f"{age:<6} {income:<10} {score:<8} {decision}")
````
