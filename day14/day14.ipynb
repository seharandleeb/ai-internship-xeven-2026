## Task 1 — Data Structures Cheat Sheet

### List vs Tuple vs Set vs Dict

| Structure | Ordered | Mutable | Duplicates | Best use case |
|---|---|---|---|---|
| **List** | Yes | Yes | Yes | An ordered, changeable sequence (a queue of items, rows of data) |
| **Tuple** | Yes | No | Yes | A fixed record that must not change (coordinates, a DB row) |
| **Set** | No | Yes | No | Unique items + fast membership tests (tags, deduplication) |
| **Dict** | Yes* | Yes | Keys must be unique | Key→value lookups (a contact book, config) |

\* `dict` preserves insertion order since Python 3.7.

### Time complexity (Big-O, average case)

| Operation | List | Tuple | Set | Dict |
|---|---|---|---|---|
| Access by index/key | O(1) | O(1) | n/a | O(1) |
| Search (`in`) | O(n) | O(n) | O(1) | O(1) (by key) |
| Insertion | O(1) append | immutable | O(1) | O(1) |
| Deletion | O(n) | immutable | O(1) | O(1) |

**Takeaway:** when you need fast membership checks or deduplication, a set or dict beats a list (O(1) vs O(n)).

The interactive menu lives in `run_menu()` / `main()` (it uses `input()`, so it runs in a terminal, not here). Below we demonstrate the underlying functions directly.

````python
import contact_manager as cm
importlib.reload(cm)

contacts = {}
cm.add_contact(contacts, 'Ali Khan', '0300-1111111',
               'ali@x.com', tags={'work', 'vip'})
cm.add_contact(contacts, 'Sara Malik', '0311-2222222',
               'sara@y.com', tags={'work', 'family'})
cm.add_contact(contacts, 'Bilal Ahmed', '0322-3333333',
               'bilal@z.com', tags={'friends'})
print('Total contacts:', len(contacts))
````

### Search (partial match via dict comprehension)

````python
matches = cm.search_contacts(contacts, 'ali')
for cid, rec in matches.items():
    print(f"#{cid}: {rec['name']} | {rec['phone']}")
# Note: 'Sara Malik' also matches because 'ali' is inside
# 'M-ali-k' — a good example of how partial match behaves.
````

### Tag management (set operations)

````python
cm.add_tag(contacts, 3, 'work')
work = cm.find_by_tag(contacts, 'work')
print('Tagged "work":', [r['name'] for r in work.values()])
cm.remove_tag(contacts, 3, 'work')
work = cm.find_by_tag(contacts, 'work')
print('After removal:', [r['name'] for r in work.values()])
````

### JSON persistence — the key challenge
Python **sets are not JSON-serializable**, so `save_to_json` converts each tag set to a sorted list, and `load_from_json` converts it back to a set and restores integer ids. Both wrap file access in `try/except` for missing or corrupt files.

````python
cm.save_to_json(contacts, 'demo_contacts.json')
reloaded = cm.load_from_json('demo_contacts.json')
print('Reloaded total:', len(reloaded))
print('Tags are a set again?',
      isinstance(reloaded[1]['tags'], set))
print('Ids are ints again?',
      all(isinstance(k, int) for k in reloaded))

import os
os.remove('demo_contacts.json')  # tidy up the demo file
````
